## Find number and pct of matches for OpenAlex ID between the publication, zenodo, and roster dataframes

```python
# Create a set of known OpenAlex IDs for fast lookup
known_ids = set(combined_teams["AllOpenAlex_ID"].dropna())

# Compare each value in PersonOpenAlex_ID to the known IDs
pub_people["AlexID_match"] = pub_people["PersonOpenAlex_ID"].apply(
    lambda x: 1 if x in known_ids else 0
)

# Count matches and non-matches
match_counts = pub_people["AlexID_match"].value_counts().sort_index()

# Calculate percentages
match_percentages = match_counts / match_counts.sum() * 100

# Print results
print("Match Counts:")
print(match_counts)

print("\nMatch Percentages:")
print(match_percentages.round(2))
```

```python
# Create a set of known OpenAlex IDs for fast lookup
known_ids = set(combined_teams["AllOpenAlex_ID"].dropna())

# Compare each value in PersonOpenAlex_ID to the known IDs
zenodo_people["AlexID_match"] = zenodo_people["PersonOpenAlex_ID"].apply(
    lambda x: 1 if x in known_ids else 0
)

# Count matches and non-matches
match_counts = zenodo_people["AlexID_match"].value_counts().sort_index()

# Calculate percentages
match_percentages = match_counts / match_counts.sum() * 100

# Print results
print("Match Counts:")
print(match_counts)

print("\nMatch Percentages:")
print(match_percentages.round(2))
```

```python
# Create a set of OpenAlex IDs from combined_teams for fast lookup
pub_tracker_ids = set(pub_people["PersonOpenAlex_ID"].dropna())

# Mark each row in combined_teams with 1 if it appears in pub_tracker_ids, else 0
combined_teams["in_pub_tracker"] = combined_teams["AllOpenAlex_ID"].apply(
    lambda x: 1 if x in pub_tracker_ids else 0
)

# Count matches and non-matches
reverse_counts = combined_teams["in_pub_tracker"].value_counts().sort_index()

# Calculate percentages
reverse_percentages = reverse_counts / reverse_counts.sum() * 100

# Print results
print("Match Counts (in_pub_tracker):")
print(reverse_counts)

print("\nMatch Percentages (in_pub_tracker):")
print(reverse_percentages.round(2))
```

```python
# Create a set of OpenAlex IDs from combined_teams for fast lookup
zenodo_ids = set(zenodo_people["PersonOpenAlex_ID"].dropna())

# Mark each row in combined_teams with 1 if it appears in zenodo_ids, else 0
combined_teams["in_zenodo"] = combined_teams["AllOpenAlex_ID"].apply(
    lambda x: 1 if x in zenodo_ids else 0
)

# Count matches and non-matches
reverse_counts = combined_teams["in_zenodo"].value_counts().sort_index()
# Calculate percentages
reverse_percentages = reverse_counts / reverse_counts.sum() * 100

# Print results
print("Match Counts (in_zenodo):")
print(reverse_counts)

print("\nMatch Percentages (in_zenodo):")
print(reverse_percentages.round(2))
```

## Roles in publication tracker

```python
# Total number of rows
total_rows = len(pub_people)

# Count where Role is missing (NaN)
empty_role_count = pub_people['Role'].isna().sum()

# Count where Role is not missing
non_empty_role_count = total_rows - empty_role_count

# Percentages
empty_role_percent = (empty_role_count / total_rows) * 100
non_empty_role_percent = (non_empty_role_count / total_rows) * 100

# Display results
print(f"Total rows: {total_rows}")
print(f"Empty Role rows: {empty_role_count} ({empty_role_percent:.2f}%)")
print(f"Non-empty Role rows: {non_empty_role_count} ({non_empty_role_percent:.2f}%)")
```

```python
# Define the list of PI-related roles to filter
main_pi_roles = [
    'Co-PI',
    'Lead PI',
    'Lead-PI'
]

# Filter pub_people for only those roles
pi_pub = pub_people[pub_people['Role'].isin(main_pi_roles)].copy()
```

```python
pi_pub["WorkOpenAlex_ID"].nunique()
```

```python
pub_tracker["OpenAlex_ID"].nunique() - pi_pub["WorkOpenAlex_ID"].nunique()
```

```python
pi_corr_pub = pub_people[pub_people['Role'].isin(main_pi_roles) | pub_people['IsCorrespondingAuthor'].eq("Yes")].copy()
```

```python
pi_corr_pub["WorkOpenAlex_ID"].nunique()
```

```python
# Find WorkOpenAlex_IDs in pi_corr_pub that have at least one row with a role in main_pi_roles
ids_with_main_pi_role = pi_corr_pub.loc[pi_corr_pub['Role'].isin(main_pi_roles), 'WorkOpenAlex_ID'].dropna().unique()

# Filter pi_corr_pub for rows where:
# - WorkOpenAlex_ID is NOT in ids_with_main_pi_role
# - Role is blank (NaN)
# - IsCorrespondingAuthor is "Yes"
filtered_rows = pi_corr_pub[
    (~pi_corr_pub['WorkOpenAlex_ID'].isin(ids_with_main_pi_role)) &
    (pi_corr_pub['Role'].isna()) &
    (pi_corr_pub['IsCorrespondingAuthor'] == "Yes")
]

filtered_rows
```

```python
filtered_rows.to_csv("works_with_only_corresponding_authors_and_not_in_the_network.csv", index=False)
```

```python
# Get WorkOpenAlex_IDs in pi_corr_pub that have at least one row with a role in main_pi_roles
ids_with_main_pi_role = pi_corr_pub.loc[pi_corr_pub['Role'].isin(main_pi_roles), 'WorkOpenAlex_ID'].dropna().unique()

# Filter pi_corr_pub for WorkOpenAlex_IDs NOT in ids_with_main_pi_role
subset = pi_corr_pub[~pi_corr_pub['WorkOpenAlex_ID'].isin(ids_with_main_pi_role)].copy()

# For each WorkOpenAlex_ID, check if all rows with IsCorrespondingAuthor == "Yes" have Role as NaN
def all_corr_no_role(group):
    corr_rows = group[group['IsCorrespondingAuthor'] == "Yes"]
    return not corr_rows.empty and corr_rows['Role'].isna().all()

valid_ids = subset.groupby('WorkOpenAlex_ID').filter(all_corr_no_role)['WorkOpenAlex_ID'].unique()

# Return only rows for those WorkOpenAlex_IDs, where IsCorrespondingAuthor == "Yes" and Role is NaN
result = subset[
    (subset['WorkOpenAlex_ID'].isin(valid_ids)) &
    (subset['IsCorrespondingAuthor'] == "Yes") &
    (subset['Role'].isna())
]

result
```

```python
# Find all WorkOpenAlex_IDs that have at least one PI or corresponding author
pi_or_corr_ids = pub_people.loc[
    pub_people['Role'].isin(main_pi_roles) | pub_people['IsCorrespondingAuthor'].eq("Yes"),
    'WorkOpenAlex_ID'
].dropna().unique()

# Remove all rows for those WorkOpenAlex_IDs
no_corr_or_pi_people = pub_people[~pub_people['WorkOpenAlex_ID'].isin(pi_or_corr_ids)].copy()
```

```python
# Find WorkOpenAlex_IDs where all rows for that ID have Role as NaN and IsCorrespondingAuthor as "No"
mask_all_role_nan = no_corr_or_pi_people.groupby('WorkOpenAlex_ID').apply(
    lambda g: g['Role'].isna().all() and (g['IsCorrespondingAuthor'] == "No").all()
)

# Get the WorkOpenAlex_IDs that meet the criteria
ids_all_role_nan_and_no_corr = mask_all_role_nan[mask_all_role_nan].index.tolist()
ids_all_role_nan_and_no_corr
```
